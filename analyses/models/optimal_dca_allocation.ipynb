# Optimal Dollar Cost Averaging (DCA) Allocation strategy

This document examines a problem I'd wondered about for a while, which is how to optimally allocate to a portfolio while dollar cost averaging.

## Problem statement

Informally, I'm curious how one should allocate to a portfolio whose composition deviates from its desired allocations between deposits, e.g., as in a monthly dollar cost averaging strategy.  In more structured terms, consider an initial portfolio allocation $P_0$, which is made of up the allocations $P_{0, j}$ for $0\le j \le n$, or $n\ge0$ separate asset allocations.  This defines an initial composition, where each asset class makes up the proportion  
$p_{0,j} := \frac{P_{0,j}}{\sum_{k=0}^n P_{0,k}}\quad .$  

Supose now that this portfolio was constructed at time $t_0$, and we can time-evolve our portfolio allocations to a future times.  For a given time $t_i$ with portfolio $P_i$ (and allocations $P_{i, j}$ for $0\le j\le n$), we can determine the portfolio at time $t_{i+1}$ via  
$P_{i+1} = G(\Delta t_{i+1}) P_{i} \quad ,$  
where  
$\Delta t_{i + 1} = t_{i + 1} - t_i \quad ,$  
and $G$ depends on the magnitude of $\Delta t_{i + 1}$ , but not the specific values of the $P_{i,j}.  It stands to reason that each asset class could evolve differently than each other (e.g., due to different rates of return), so $G$ is allowed to depend on the class itself, but again not on the particular values of $P_{i, j}$.  In other words, $G$ has a matrix representation, so that  
$P_{i+1,k} = \sum_{j = 0}^n G_{k, j}(\Delta t_{i + 1}) P_{i, j} \quad . $  
Further, if $G$ is not dependent on the particular values in $P_{i,j}$, then we can construct the portfolio at time $t_i$, $i\ge 1$, via repeated application of $G$:  
$P_{i} = \prod_{k = 0}^i G(\Delta t_k) P_0 \quad . $  

Finally, then, consider that we wish to make deposits into this portfolio at each of these future intervals, so that we consider a deposit $d_i$ at the time $t_i$.  The question can now be posed as, what is the _optimal_ allocation of each deposit $d_i$ (i.e., the components $d_{i, j}$ into each of the $n$ asset classes), given the values of the portfolio $P_i$ prior to the deposit.
