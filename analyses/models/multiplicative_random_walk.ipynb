# Multiplicative random walk

These are my notes on an interesting idea I heard on the Rational Reminder podcast; I believe the guest was Hendrik Bessembinder.

## Basic premise

To get an idea of the interesting tail of outcomes for individual stocks vs. the market return, consider the asymmetry in a single stock gaining or losing a fixed percentage in two consecutive periods.  E.g., say a stock is initially valued at $P_0$, and can either gain or lose 10% in two consecutive, independent periods.  Since $0.9 \times 1.1 = 0.99 \neq 1$, we get skewed outcomes, where the stock could be at $0.81 P_0$ with a 1/4 probability, at $0.99 P_0$ with 1/2 probability, or at $1.21 P_0$ with 1/4 probability; the mean, however, is preserved at $P_0$.  This is a multiplicative random walk,   
$P_n = \prod_{j=1}^n Z_j P_0 \quad ,$  
where the $Z_j = 1 \pm 0.1$ are independent random variables.

Naturally, one should express skepticism at modeling stock returns over consecutive periods through independent (and so necessarily _uncorrelated_) random variables, but this is a simplifying assumption.

## Basic properties

To generalize, consider $n\ge 1$ periods (the $n=0$ case is trivially pinned at $P_0$),  
$Z_j = 1\pm\delta\,\quad \quad 0 < \delta < 1 \quad ,$  
where $1 \le j \le n$ and the $Z_j$ take on each of the two values with equal probability (1/2).  As independent random variables, the expectation value of the product is given by  
$E\left( \prod_{j=1}^n Z_j \right) = \prod_{j=1}^n  E\left( Z_j \right) = 1\quad,$  
or,  
$E(P_n) = P_0 \quad .$  
To understand the variance of this multiplicative random walk, we can consider that for any $1 \le j \le n$:  
$E(Z_j^2) = \frac{1}{2}\left[ \left(1 + \delta \right)^2 + \left(1 - \delta \right)^2 \right] = 1 + \delta^2 \quad.$  
The variance of $P_n$ is then  
$E\left( P_n^2 \right) - E^2\left( P_n \right) = E\left( P_0^2\prod_{j=1}^n Z^2_j \right) - P_0^2 = P_0^2\left[ \left( 1 + \delta^2 \right)^n - 1 \right] \quad .$  

For small $\delta^2$ (which is not unrealistically restrictive, considering $0 < \delta < 1$), the variance still initially scales linearly in $n$, like the typical random walk.  
$E\left( P_n^2 \right) - E^2\left( P_n \right) = P_0^2\left[ n \delta^2 + \mathcal O\left( \delta^4 \right) \right] \quad .$  
The caveat, of course, is that these "higher order" terms in $\delta^2$ are actually exponential in $n$, and will eventually dominate for a sufficiently large $n$.  Curiously as well, this eventual exponential growth in the variance must come from exponentially large skewed outcomes, as $P_n$ can only ever approach $0$ from above.
