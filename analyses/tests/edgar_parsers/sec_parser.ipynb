## Test of historical $MSFT data

### Import an old filing

Looking at a recent 10-Q, since I confirmed this is the same kind of file returned by the [sec_downloader package](https://github.com/Elijas/sec-downloader).

```python
with open('../../../../../../.config/edgar/user-agent', 'r') as file:
    user_agent = file.read().strip()
headers = {
    'User-Agent': user_agent
}

filing = requests.get(url='https://www.sec.gov/Archives/edgar/data/789019/000095010994000252/0000950109-94-000252.txt', headers = headers)
filing.status_code
```

```python
elements = sp.Edgar10QParser().parse(filing.content)
tree = sp.TreeBuilder().build(elements)
tree_output = sp.render(tree)

# Test function from the example documentation
def print_first_n_lines(text: str, *, n: int):
    print("\n".join(text.split("\n")[:n]), "...", sep="\n")

print_first_n_lines(tree_output, n=30)
```

### result: failure

If we compare this tree output to the example in [the README's usage](https://github.com/alphanome-ai/sec-parser#usage) section, we're missing a lot of the fine-grained detail.  The parser has parsed the text (I presume), but not the table structure (the linked example refers to a `TableElement` missing in the above output).  That said, it does at least extract the main part titles correctly ("Financial Information" and "Other Information").
