# Dividends vs. stock buybacks

I've been rather interested in how the market--a suitably broad index of American stocks--grows over time, and the remarkably sustainable market performance over around 30 years.  Since dividend re-investment is such a fundamental aspect of long-term investing, I'm naturally drawn to the question of whether the current era of stock buybacks could have long-term effects on investment returns.  The appeal of stock buybacks is clear: deferred taxes until you actually close your position, at which point the relevant tax penalties are based on short or long term capital gains.  I wonder, however, if the more nebulous aspects are really understood.  E.g.,:  
- Do stock buybacks manipulate buy signals to drive prices higher, or is there an instrinsic reason for prices to rise?
- Does the value delivered to investors outweigh the tax penalty on dividends?
- Does the market effectively price in stock buybacks?  If not, why?
- Don't stock buybacks overly concentrate an investors portfolio?  Wouldn't the dividend be preferable as diversifying with up-to-date information?

I'm going to test out various analyses in what follows, first by assuming the efficient market hypothesis (suitably defined) and later allowing for mispricing.  Eventually, I'd like to backtest strategies based on dividendifying buybacks, but that may have to wait for further analysis.

## A comparison assuming efficient market pricing

In what follows, I consider a scenario where a company returns a dividend to shareholders and a separate scenario where a company executes a share buyback.  I'm assuming that the this process can be understood from market data established suitably before and suitably after the divided or buyback; if the announcement is crucial to the market pricing, then I include it within the dividend/buyback process.  In particular, I will consider the market capitalization $M$,  
$M_{i,f} = s_{i,f} P_{i,f} \,\,,$  
where the $(i,f)$ subscripts indicate values before or after the return to investors, $s$ is the number of shares outstanding, and $P$ is the price per share.  I'll denote the amount returned to investors--through either mechanism--as $d$.

By "efficient market pricing," I mean that the market capitalization before executing a return to shareholders essentially prices in this return.  I.e., the return is predicted sufficiently soon that the present value of the predicted payout more or less mirrors the actual amount returned to shareholders.  Specifically, I'm assuming that we can choose times before and after the return (in either mechanism) such that  
$M_f = M_i - d \,\, .$  
Such a simple relation obviously misses (important) details like the after-tax return to shareholders, but--as minor foreshadowing--I don't believe this discrepancy affects the conclusion.

### Dividend payouts

I'm assuming that there are no (appreciable) changes in the number of shares outstanding over the course of the dividend payout.  The final stock price after the payout then  
$P_f = \frac{M_f}{s_f} = \frac{M_i - d}{s_i} = P_i - \frac{d}{s_i} \,\,\,,$  
where the remainder is the dividend payout per share.  In other words, the price per share goes down by the amount paid out per share, which is the expected balancing act for an efficiently priced shareholder payout.

### Stock buybacks

Perhaps controversially (or incorrectly), I'm assuming the same change in market capitalization before and after the buyback ($M_f = M_i - d$).  My reasoning is that, if the company has $d$ to payout to shareholders, then I'm analyzing the expense as that.  If the company buys shares speculatively, then I wouldn't consider that a return to shareholders, but some sort of company financial activity, which itself may be worth further shareholder discussion.  In particular, the amount "returned" would necessarily be less than the amount spent ($d$) after the re-issuance of the shares.  Alternatively, the company might buy shares as a hedge against a payout.  If things are going well, the treasury stock could be re-issued to raise capital or as incentives to employees.  If things go poorly, the treasury stock could be retired and marked down as a return to shareholders.  In both of these latter cases, I would have preferred an initial dividend payout.

Now that I've clarified my bias, I'm assuming the basic rational investor assumption that the shares are bought back at a price $P_b$  greater than or equal to the market price $P_i$.  If bought on the open market, I could imagine variations in the purchase price that might justify considering the weighted average  
$P_b = \left(\frac{1}{s_b}\right)\sum_{j} P_{b,j} s_{b, j} \,\,,$  
where $s_{b,j}$ denotes the number of shares purchased at the price $P_{b,j}$ and $s_b$,  
$s_b = \sum_j s_{b,j}\,\,,$  
is the total number of shares purchased in the buyback.  Assuming efficient (or rational, etc.) market behavior, we have not only that $P_b\ge P_i$, but that each market repurchase transaction must be in the investor's interest, or $\forall j: P_{b,j} \ge P_i$.  If the company spends $d$ in total buybacks, then  
$d = s_b P_b\,\,.$

After the buybacks, the total shares oustanding become  
$s_f = s_i - s_b\,\,.$  
The price per share is now  
$P_f = \frac{M_f}{s_f} = \frac{M_i - d}{s_i - s_b} = \frac{M_i/s_i - d/s_i}{1 - s_b/s_i}
= \frac{P_i - P_b s_b/s_i}{1 - s_b/s_i} <\frac{P_i - P_i s_b/s_i}{1-s_b/s_i} = P_i \,\,,$  
where the relationships $d=s_bP_b$ and the restriction $P_b \ge P_i$ were used in intermediary substitutions.  I can imagine a combination of scales $d$ and $s_b$ where this value exceeds $P_i$, but the efficient pricing hypothesis restricts $P_f$ to be _less_ than the initial price.  As a result, stock buybacks of an efficiently priced stock _should_ presumably reduce the stock price with no compensatory payout.

### Buying back undervalued stock

One argument I've seen is that stock buybacks make sense when the stock is undervalued.  To make this concrete, let's relate the company's market capitalization $M$ to the present value of future dividends $V$ plus an error term, $\epsilon$.  
$M = V + \epsilon$  
The stock buyback will cost $C_B$ in present-day dollars,  
$C_B = s_B P_B$,  
where $s_B$ is the total shares purchased in the buyback and $P_B$ is the price per share.  If the shares are purchased at multiple prices during the buyback, $P_B$ can equally represent the weighted average, or mean purchase price per share.  As a result of the buyback, the company's intrinsic value must necessarily decrease by the cost (otherwise, this is not a mechanism of returning value to shareholders, but some sort of corporate financing which should be separately scrutinized through that lens).  Therefore,  
$V_f = V_i - C_B$.  
Naturally, the total stock outstanding must decrease by the amount repurchased in the buyback, or  
$s_f = s_i - s_B$.

I'm defining the stock buyback as _worthwhile_ if the intrinsic value (present value of future dividends) per share, $V/s$, is increased as a result, or  
$\frac{V_i}{s_i} \le \frac{V_f}{s_f}$.   
Using the previous relations, we have  
### $\frac{V_i}{s_i} = P_i - \frac{\epsilon_i}{s_i} \le \frac{s_i P_i - s_B P_B - \epsilon_i}{s_i - s_B} = \frac{V_f}{s_f}$  
Shuffling terms around, the inner inequality can be written as  
#### $P_B - P_i \le -\frac{\epsilon_i}{s_i}$.  
In other words, leftover shareholder value can only be increased through a buyback if the stock price-per-share is undervalued by an amount $-\epsilon_i/s_i>0$, which exceeds the difference in the market price $P_i$ and the re-purchase price $P_B$.  Notably, this is the threshold criteria; the actual value per share increase is  
### $\frac{V_f}{s_f} - \frac{V_i}{s_i} = \frac{s_B}{s_i - s_B}\left[ \frac{-\epsilon_i}{s_i} - \left( P_B - P_i \right) \right]$.  
As a reminder, $-\epsilon_i >0$ corresponds to an initially _undervalued_ stock, with an undervaluation of $-\epsilon_i/s_i$ per share.
